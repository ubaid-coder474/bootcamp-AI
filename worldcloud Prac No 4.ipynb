```python
# Import necessary libraries
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter

# Ensure necessary NLTK data is downloaded
nltk.download('punkt')
nltk.download('stopwords')
```

```python
text = """
Football is one of the most popular sports worldwide. It is played by two teams of eleven players each. The objective of the game is to score more goals than the opposing team by getting the ball into the opponent's goal. The game is played on a rectangular field with goalposts at either end. The players use their feet to kick the ball, and they also use their heads, chests, and knees. The team that scores the most goals during the match wins.

"""
print(text)
```

```python
words = word_tokenize(text.lower())  # Convert to lowercase for uniformity
print(words)
```

```python
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.isalpha() and word not in stop_words]
print(filtered_words)
```

```python
word_freq = Counter(filtered_words)
print(word_freq)
```

```python
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)
```

```python
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off') 
plt.show()
```
